```python
import pandas as pd 
```

```python
df=pd.read_csv("C:\\Users\\tusha\\OneDrive\\Desktop\\heart\\HEART_DISEASE_PREDICTION\\Data\\HEART.csv")
```

```python
X=df.drop("target",axis=1)
y=df["target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
ranforclf=RandomForestClassifier()
```

```python
ranforclf.fit(X_train,y_train)
```

```python
y=ranforclf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y,y_test)
```

```python
parameters={
    "n_estimators":[100,200,300],
    "criterion":["gini","entropy","log_loss"],
    "max_depth":[1,2,3,None],
    "min_samples_split":[1,2,3,4],
    "max_features":["sqrt","log2",None]
}
```

```python
gclf=GridSearchCV(ranforclf,param_grid=parameters,cv=3,scoring="accuracy",verbose=3)
```

```python
gclf.fit(X_train,y_train)
```

```python
gclf.best_params_
```

```python
yclf=gclf.predict(X_test)
```

```python
accuracy_score(yclf,y_test)
```

```python
rclf=RandomForestClassifier(criterion='entropy',max_depth= 3,max_features="sqrt",min_samples_split= 2,n_estimators= 100)
```

```python
rclf.fit(X_train,y_train)
```

```python
y_pred=rclf.predict(X_test)
```

```python

accuracy_score(y_pred,y_test)
```
